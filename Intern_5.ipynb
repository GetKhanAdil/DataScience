# 

 Merging the Matches and Delivery csv file

```python
Matches.head(1)
```

```python
seas=Matches[['id','season','winner']]
```

```python
complete=Delivery.merge(seas,how="inner",left_on='match_id',right_on='id')
complete.shape
```

```python
complete.tail()
```

```python
comp_4=complete[complete.batsman_runs==4]
comp_6=complete[complete.batsman_runs==6]
```

```python
season_4=pd.pivot_table(index='season',values='batsman_runs',aggfunc="sum",data=comp_4).sort_values(by='season',ascending=False)
season_4["fours"]=season_4.batsman_runs/4
```

```python
a=season_4.drop('batsman_runs',axis=1)
a
```

```python
a.plot(kind="bar",figsize=(12,8),title="No. of Fours hit",grid=True)
```

 In 2013 max No of 4's hit (2054)

```python
season_6=pd.pivot_table(index='season',values='batsman_runs',aggfunc="sum",data=comp_6).sort_values(by='season',ascending=False)
season_6["six"]=season_6.batsman_runs/6
season_6
```

```python
b=season_6.drop('batsman_runs',axis=1)
b.plot(kind="bar",figsize=(12,8),title="Total No. of Six in each Season ",grid=True,color="g")
```

Most sixes hit in 2018 with total count of 5214

```python
Runs=pd.pivot_table(index='batsman',values='batsman_runs',aggfunc="sum",data=complete).sort_values(by='batsman_runs',ascending=False)
Runs[:10]
```

```python
Runs[:10].plot(kind="bar",figsize=(12,8),title="LEADING RUN SCORER ",grid=True,color="y")
```

SURESH RAINA & VIRAT KOHLI ARE LEADING RUN SCORER WITH RUN OF 5434  , 5415
