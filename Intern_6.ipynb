### Importing the required Libraries

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm

import pandas as pd
import numpy as np
import seaborn as sns

from sklearn import tree
import matplotlib.pyplot as plt
```

# 

### Loading Data set

```python
iris = load_iris()
X=iris.data[:,:] 
y=iris.target
```

#### Exploratory Data Analysis

```python
data=pd.DataFrame(iris['data'],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data['Species']=iris['target']
data['Species']=data['Species'].apply(lambda x: iris['target_names'][x])

data.head()
```

```python
data.shape
```

```python
data.describe()
```

#### Data Visualization

```python
sns.pairplot(data)
```

# 

```python
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Sepal Length','Sepal Width').add_legend()
plt.show()

# Scatter plot of data based on Petal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Petal length','Petal Width').add_legend()
plt.show()
```

### Decision Tree Model Training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) 
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train,y_train)
print("Training Complete.")
y_pred = tree_classifier.predict(X_test)
```

#### Comparing the actual and predicted flower classification

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df.head()
```

### Visualizing the Trained Model

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(tree_classifier, feature_names = fn, class_names = cn, filled = True);
```

```python
tree_classifier.score(X_test,y_test)
```
