### Removing the Error value

```python
amazon2=amazon1[~amazon1.retail_price_amaz<0]
```

```python
amazon3=amazon2[~amazon2.saving_amaz<0]
```

## merging the flipkart and amazon with index

```python
df = pd.merge(flip2, amazon3,left_index=True, right_index=True)
```

```python
df.shape
```

```python
df.head(2)
```

```python
plt.figure(figsize=[15,10])
sns.histplot(df.discount_percent_flip,color='green',label="flipkart",kde=True)
sns.histplot(df.discount_percent_amaz,color='red',label="Amazon",kde=True)
plt.legend()
plt.show()
```
