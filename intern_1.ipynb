### In This Small Project We Are Going To Predict The Score Of a Student Based On  No. Of Hours They Studied.
### Here We Are Using LinearRegreesion Algorithm To Predict The Scores.


```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("quest.csv")
df.head()
```

# plotting the data to show any type of relationship between them

```python
plt.figure(dpi=150)
plt.scatter(df.Hours,df.Scores,marker='*')
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=[15,10])
sns.barplot(x='Hours',y='Scores',data=df)
plt.show()
```

Reshaping the data to make it 2D array

```python
df1=np.array(df.Hours)
X1=df1.reshape(-1,1)
```

```python
df2=np.array(df.Scores)
Y1=df2.reshape(-1,1)
```

Spliting the data into traning and testing phase for traning and testing

```python
from sklearn.model_selection import train_test_split
train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X1,Y1,test_size=.3)
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
print("Traning Complete")
```

```python
pred=lr.predict(x_test)
print("Predicted score",pred)
```

```python
lr.intercept_
print("intercept ==>",lr.intercept_)
```

```python
lr.coef_
```

```python
lines=lr.coef_*X1+lr.intercept_
```

```python
plt.figure(dpi=125)
plt.title("Plotting The Regression Line")
plt.scatter(X1,Y1,marker='+',label="data spread")
plt.plot(X1,lines,label='Regression Line',color='g')
plt.legend()
plt.show()
```

```python
y_pred_9=lr.predict([[9.5]])
```
